# Conversation Token Buffer Memory
Limit by token size (a size of a conversation)

```python
from langchain_openai import ChatOpenAI
from langchain.memory import ConversationTokenBufferMemory
from langchain.chains.conversation.base import ConversationChain

llm = ChatOpenAI()
memory = ConversationTokenBufferMemory(llm=llm, max_token_limit=250)

conversation_chain = ConversationChain(
    llm=llm,
    memory=memory,
    verbose=True
)

conversation_chain.predict(input="Hello, my name is Vinicius.")
conversation_chain.predict(input="Do you know my name?")
```
