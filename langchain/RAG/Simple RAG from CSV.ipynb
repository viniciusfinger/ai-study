# Simple RAG from CSV

```python
from langchain_community.document_loaders.csv_loader import CSVLoader

path = 'documents/Top 1000 IMDB movies.csv'
loader = CSVLoader(path)

content = loader.load()

print(content[0].page_content) # Linha 0
print("Metadata: " + content[0].metadata)
```

### Fazendo perguntas pro CSV

```python
from langchain.chains.question_answering import load_qa_chain
from langchain.chat_models import ChatOpenAI

llm = ChatOpenAI(model="gpt-3.5-turbo")

chain = load_qa_chain(llm=llm, chain_type="stuff", verbose=True) # stuff é um tipo de retrieval

chain.run(
    input_documents=content[:100],
    question="Qual o filme mais recente a entrar na lista?"
)
```

```python
chain.run(
    input_documents=content[:100],
    question="Qual o filme mais antigo da lista?"
)
```
