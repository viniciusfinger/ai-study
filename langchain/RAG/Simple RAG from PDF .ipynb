# Simple RAG from PDF
Para PDFs pequenos, é possível fazer RAG em um pdf sem passar por embedding e persistir em um vector store:


```python
from langchain_community.document_loaders.pdf import PyPDFLoader

path = 'guia_alimentar_populacao_brasileira.pdf'
loader = PyPDFLoader(path)

pdf_pages = loader.load()

# Printa uma determinada página
print(pdf_pages[0])
```

### Fazendo perguntas a esse documento
Em documentos pequenos em que sabemos o conteúdo, é possível fazer perguntas ao documento sem precisar de um vector store:

```python
from langchain.chains.question_answering import load_qa_chain
from langchain.chat_models import ChatOpenAI

llm = ChatOpenAI(model="gpt-3.5-turbo")

chain = load_qa_chain(llm=llm, chain_type="stuff", verbose=True) # stuff é um tipo de retrieval

print(chain.run(
    input_documents=pdf_pages[:10],
    question="Quando esse livro foi publicado orignialmente?"
))
```

```python
print(chain.run(
    input_documents=pdf_pages[:10],
    question="O que é a Lei Orgânica de Segurança Alimentar e Nutricional?"
))
```
